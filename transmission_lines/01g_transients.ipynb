```python
from IPython.html.widgets import interact, fixed
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
from IPython.display import display, Math, Latex
from matplotlib import animation
from JSAnimation import IPython_display
```

# Set transmission line values

```python
v_g = 1.0
Z_0 = 50.0
Z_L = 150.0
Z_g = 25.0
loadreflcoeff = (Z_L - Z_0)/(Z_L + Z_0)
genreflcoeff = (Z_g - Z_0)/(Z_g + Z_0)
v1_plus = v_g * Z_0 / (Z_g + Z_0)
v_steadystate = v_g * float(Z_L) / (Z_L + Z_g)

tline_length = 30.0

nbounces = 10
```

# Display transmission line parameters

```python
tempstr = 'v_g = %.3f V' % v_g; display(Math(tempstr))
tempstr = 'Z_0 = %.3f \Omega' % Z_0; display(Math(tempstr))
tempstr = 'Z_L = %.3f \Omega' % Z_L; display(Math(tempstr))
tempstr = 'Z_g = %.3f \Omega' % Z_g; display(Math(tempstr))
print '---------------------'
tempstr = '\Gamma_L = %.3f' % loadreflcoeff; display(Math(tempstr))
tempstr = '\Gamma_g = %.3f' % genreflcoeff; display(Math(tempstr))
print '---------------------'

# Print values for v1+, v1-, v2+, v1-, etc.
def print_delta_voltages(nbounces, v1_plus, loadreflcoeff, genreflcoeff):
    subscript_index = 1
    delta_voltages = np.zeros(nbounces)
    delta_voltages[0] = v1_plus
    tempstr = '$v_{%d}^+ = %.5f V' % (subscript_index, delta_voltages[0])
    display(Math(tempstr))
    for i in range(1,nbounces):
        if i%2 == 0:
            delta_voltages[i] = delta_voltages[i-1] * genreflcoeff
            subscript_index += 1
            tempstr = '$v_{%d}^+ = %.5f V' % (subscript_index, delta_voltages[i])
        else:
            delta_voltages[i] = delta_voltages[i-1] * loadreflcoeff
            tempstr = '$v_{%d}^- = %.5f V' % (subscript_index, delta_voltages[i])
        display(Math(tempstr))

print_delta_voltages(nbounces, v1_plus, loadreflcoeff, genreflcoeff)

print '---------------------'

tempstr = 'v_{steadystate} = %.3f V' % v_steadystate; display(Math(tempstr))
```

# Display voltage parameterized by t/T

```python
numpnts_in_plot = 101

# Set up array with v1+, v1-, v2+, v1-, etc.
def calc_delta_voltages(nbounces, v1_plus, loadreflcoeff, genreflcoeff):
    delta_voltages = np.zeros(nbounces)
    delta_voltages[0] = v1_plus
    for i in range(1,nbounces):
        if i%2 == 0:
            delta_voltages[i] = delta_voltages[i-1] * genreflcoeff
        else:
            delta_voltages[i] = delta_voltages[i-1] * loadreflcoeff
    return delta_voltages

z = np.linspace(0,tline_length,num=numpnts_in_plot)
v = np.zeros(numpnts_in_plot)

def calc_voltage_on_tline(t_normalized,reflcoeff_load,reflcoeff_gen):
    global v
    nbounces, fractalongtline = divmod(t_normalized,1)
    nbounces_int = int(nbounces)
    delta_voltages = calc_delta_voltages(nbounces_int+1, v1_plus, loadreflcoeff, genreflcoeff)
    basevoltage = 0.0
    if nbounces_int != 0:
        for i in range(nbounces_int):
            basevoltage += delta_voltages[i]
    v[:] = basevoltage
    next_increment = delta_voltages[nbounces_int]
    if nbounces_int % 2 == 0:
        index = round(fractalongtline*numpnts_in_plot)
        v[:index] += next_increment
    else:
        index = round((1.0 - fractalongtline)*numpnts_in_plot)
        v[index:] += next_increment

def plot_voltage_on_tline(t_normalized,reflcoeff_load,reflcoeff_gen):
    calc_voltage_on_tline(t_normalized,reflcoeff_load,reflcoeff_gen)
    text1 = 'v_steadystate = %.3f V\n' % v_steadystate
    text1 += '$\Gamma_L$ = %.3f\n' % loadreflcoeff
    text1 += '$\Gamma_g$ = %.3f' % genreflcoeff
    text2 = 'v_g = %.3f V\n' % v_g
    text2 += '$Z_0$ = %.1f\n' % Z_0
    text2 += '$Z_L$ = %.1f\n' % Z_L
    text2 += '$Z_g$ = %.1f' % Z_g
    text3 = 't = %.2f T' %t_normalized
    fig = plt.figure(figsize=(10,6))
    plt.plot(z,v)
    plt.ylim(0,1.5*v_g)
    plt.xlabel('z')
    plt.ylabel('voltage')
    plt.text(2,1.05*v_g,text2)
    plt.text((float(tline_length)/2),1.1*v_g,text1)
    plt.text(2, 0.2, text3)

interact(plot_voltage_on_tline, t_normalized=(0.0,6.0), 
         reflcoeff_load=fixed(loadreflcoeff), reflcoeff_gen=fixed(genreflcoeff))
```

# Animation of voltage on transmission line

```python
numpnts_in_plot = 101

# Set up array with v1+, v1-, v2+, v1-, etc.
def calc_delta_voltages(nbounces, v1_plus, loadreflcoeff, genreflcoeff):
    delta_voltages = np.zeros(nbounces)
    delta_voltages[0] = v1_plus
    for i in range(1,nbounces):
        if i%2 == 0:
            delta_voltages[i] = delta_voltages[i-1] * genreflcoeff
        else:
            delta_voltages[i] = delta_voltages[i-1] * loadreflcoeff
    return delta_voltages

z = np.linspace(0,tline_length,num=numpnts_in_plot)
v = np.zeros(numpnts_in_plot)

def calc_voltage_on_tline(t_normalized,reflcoeff_load,reflcoeff_gen):
    global v
    nbounces, fractalongtline = divmod(t_normalized,1)
    nbounces_int = int(nbounces)
    delta_voltages = calc_delta_voltages(nbounces_int+1, v1_plus, loadreflcoeff, genreflcoeff)
    basevoltage = 0.0
    if nbounces_int != 0:
        for i in range(nbounces_int):
            basevoltage += delta_voltages[i]
    v[:] = basevoltage
    next_increment = delta_voltages[nbounces_int]
    if nbounces_int % 2 == 0:
        index = round(fractalongtline*numpnts_in_plot)
        v[:index] += next_increment
    else:
        index = round((1.0 - fractalongtline)*numpnts_in_plot)
        v[index:] += next_increment

numframes = 161
T_numframes = numframes/8
zmin = 0.0
zmax = tline_length
numpnts = numpnts_in_plot
u = float(zmax - zmin)/numframes

text1 = 'v_steadystate = %.3f V\n' % v_steadystate
text1 += '$\Gamma_L$ = %.3f\n' % loadreflcoeff
text1 += '$\Gamma_g$ = %.3f' % genreflcoeff
text2 = 'v_g = %.3f V\n' % v_g
text2 += '$Z_0$ = %.1f\n' % Z_0
text2 += '$Z_L$ = %.1f\n' % Z_L
text2 += '$Z_g$ = %.1f' % Z_g

# First set up the figure, the axis, and the plot element we want to animate
fig = plt.figure(figsize=(10,6))
ax = plt.axes(xlim=(zmin, zmax), ylim=(0,1.5*v_g))
ax.set_xlabel('z (m)')
ax.set_ylabel('Voltage (V)')
ax.text(0.1, 0.75, text2, transform = ax.transAxes)
ax.text(0.6, 0.75, text1, transform = ax.transAxes)
line, = ax.plot([], [], lw=2)
textfortime = ax.text(0.1, 0.15, '', transform = ax.transAxes)

# initialization function: plot the background of each frame
def init():
    line.set_data([], [])
    text3 = ''
    textfortime.set_text(text3)
    return line, textfortime,

# animation function.  This is called sequentially
def plot_v(t):
    global v
    z = np.linspace(zmin,zmax,numpnts)
    v = np.zeros(numpnts)
    t_normalized = float(t)/T_numframes
    calc_voltage_on_tline(t_normalized, loadreflcoeff, genreflcoeff)
    text3 = 't = %.2f T' % t_normalized
    textfortime.set_text(text3)
    line.set_data(z,v)
    return line, textfortime,

# call the animator.  blit=True means only re-draw the parts that have changed.
animation.FuncAnimation(fig, 
                               plot_v, 
                               init_func=init,
                               frames=numframes, 
                               interval=20, 
                               blit=True
                               )
```
